Simple Plots

```python
from mpmath import mpf, mp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mp.dps = 3
# load data
filepath = 'data/roots_130.csv'
roots = pd.read_csv(filepath)
```

```python
roots
```

```python
xstrings = roots['real'].to_numpy()
ystrings = roots['imaginary'].to_numpy()
qstrings = roots['q'].to_numpy()
errorstrings = roots['error'].to_numpy()
xs = []
ys = []
qs = []
es = []
for x in xstrings:
    if x not in roots.columns:
        xs.append(mpf(x))
for y in ystrings:
    if y not in roots.columns:
        ys.append(mpf(y))
for q in qstrings:
    if q not in roots.columns:
        qs.append(float(q))
for e in errorstrings:
    if  e not in roots.columns:
        es.append(mpf(e))
```

```python
fig, axs = plt.subplots(1,figsize=(20,20))
axs.scatter(xs,ys, c=qs, s=5)
plt.xlim(-200,200)
plt.ylim(-200,200)
```

```python
fig, axs = plt.subplots(1,figsize=(20,20))
axs.scatter(xs,ys, c=qs, s=5)
plt.xlim(-1,1)
plt.ylim(-1,1)
```
