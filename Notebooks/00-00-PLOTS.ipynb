Simple Plots

```python
from mpmath import mpf, mp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mp.dps = 3
# load data
filepath = '../data/roots_200.csv'
roots = pd.read_csv(filepath)
```

```python
#load up the discriminant roots
d_roots = pd.read_csv('../data/discriminant_roots_100.csv')
d_roots
```

```python
roots
```

```python
xstrings = roots['real'].to_numpy()
ystrings = roots['imaginary'].to_numpy()
qstrings = roots['q'].to_numpy()
pstrings = roots['p'].to_numpy()
xs = []
ys = []
qs = []
es = []
ps = []
for x in xstrings:
    if x not in roots.columns:
        xs.append(mpf(x))
for y in ystrings:
    if y not in roots.columns:
        ys.append(mpf(y))
for q in qstrings:
    if q not in roots.columns:
        qs.append(float(q))
for p in pstrings:
    if p not in roots.columns:
        ps.append(float(p))
```

```python
# repeat for the discriminant
d_xstrings = d_roots['real'].to_numpy()
d_ystrings = d_roots['imaginary'].to_numpy()
d_qstrings = d_roots['q'].to_numpy()
d_pstrings = d_roots['p'].to_numpy()
d_xs = []
d_ys = []
d_qs = []
d_es = []
d_ps = []
for x in d_xstrings:
    if x not in d_roots.columns:
        d_xs.append(mpf(x))
for y in d_ystrings:
    if y not in d_roots.columns:
        d_ys.append(mpf(y))
for q in d_qstrings:
    if q not in d_roots.columns:
        d_qs.append(float(q))
for p in d_pstrings:
    if p not in d_roots.columns:
        d_ps.append(float(p))
```

```python
ratio = np.array(ps)/np.array(qs)
```

```python
fig, axs = plt.subplots(1,figsize=(20,20))
s = axs.scatter(xs,ys, c=qs, s=1)
lim = 1
plt.xlim(-lim,lim)
plt.ylim(-lim,lim)
fig.colorbar(s)
```

```python
# color by ratio
fig, axs = plt.subplots(1,figsize=(20,20))
s = axs.scatter(xs,ys, c=ratio, s=5)
lim = 1
plt.xlim(-lim,lim)
plt.ylim(-lim,lim)
fig.colorbar(s)
# plot discriminant roots in red
axs.scatter(d_xs,d_ys,c='r', s=3)
axs.set_title("Colored by the decimal value of p/q")
```
